```python
# Merge DNA and lipid bilayer structures using MDAnalysis
# Input: DNA.gro, lipids.gro
# Output: DNA_lipids.gro


import MDAnalysis as mda

dna_alone = "DNA.gro"
lipids_clean = "lipids.gro"
out = "DNA_lipids.gro"

u_dna = mda.Universe(dna_alone)
u_bilayer = mda.Universe(lipids_clean)

merged = mda.Merge(u_dna.atoms, u_bilayer.atoms)
merged.atoms.write(out)
```
