```python
import MDAnalysis as mda 
```

```python
u=mda.Universe("DNA_lipids.gro")
```

```python
chain1_N1_N3 = u.select_atoms("((resname DA DG and name N1) or (resname DT DC and name N3) or (resname HFA HFG HSA HSG and name N1) or  (resname HFT HFC HST HSC  and name N3)) and (resid 1:147)")
chain2_N1_N3 = u.select_atoms("((resname DA DG and name N1) or (resname DT DC and name N3)) and (resid 148:294)")
print(f"{len(chain1_N1_N3)}, {len(chain2_N1_N3)}")
```

```python
import pandas as pd

# Convert chain2_N1_N3 to a DataFrame
chain2_N1_N3_df = pd.DataFrame({
    'Atom ID': chain2_N1_N3.ids,
    'Atom Name': chain2_N1_N3.names,
    'Residue Name': chain2_N1_N3.resnames,
    'Residue ID': chain2_N1_N3.resids,
    'Segment ID': chain2_N1_N3.segids,
    'Position': list(chain2_N1_N3.positions)
})

print(chain2_N1_N3_df)
```

```python

# Convert chain2_N1_N3 to a DataFrame
chain1_N1_N3_df = pd.DataFrame({
    'Atom ID': chain1_N1_N3.ids,
    'Atom Name': chain1_N1_N3.names,
    'Residue Name': chain1_N1_N3.resnames,
    'Residue ID': chain1_N1_N3.resids,
    'Segment ID': chain1_N1_N3.segids,
    'Position': list(chain1_N1_N3.positions)
})

print(chain1_N1_N3_df)
```

```python
# Reverse chain2_N1_N3_df based on its index
chain2_N1_N3_df_reversed = chain2_N1_N3_df.iloc[::-1].reset_index(drop=True)

print(chain2_N1_N3_df_reversed)
```

```python
merged_df = pd.DataFrame({
    'Atom ID (chain1)': chain1_N1_N3_df['Atom ID'],
    'Atom ID (chain2)': chain2_N1_N3_df_reversed['Atom ID'],
    'Residue Name (chain1)': chain1_N1_N3_df['Residue Name'],
    
    'Residue Name (chain2)': chain2_N1_N3_df_reversed['Residue Name'],
    'Residue ID (chain1)': chain1_N1_N3_df['Residue ID'],
    'Residue ID (chain2)': chain2_N1_N3_df_reversed['Residue ID']
   
})
#print(merged_df)
merged_df.to_csv('merged_df.csv', index=False)
merged_df
```

```python
# Add new columns with default values
merged_df['atom_i'] = merged_df['Atom ID (chain1)']
merged_df['atom_j'] = merged_df['Atom ID (chain2)']
merged_df['func'] = 10  
merged_df['r0'] = 0.270  
merged_df['r1'] = 0.299  #
merged_df['r2'] = 0.330 # 
merged_df['k(kJ mol^-1 nm^-2)'] = 1000  # Example default value
merged_df[";"] = ";"  # Adding a column with a semicolon as its value
merged_df['Atom ID (chain1)'] = chain1_N1_N3_df['Atom ID']
merged_df['Atom ID (chain2)'] = chain2_N1_N3_df_reversed['Atom ID']
merged_df['Residue Name (chain1)'] = chain1_N1_N3_df['Residue Name']
merged_df['Residue Name (chain2)'] = chain2_N1_N3_df_reversed['Residue Name']
merged_df['Residue ID (chain1)'] = chain1_N1_N3_df['Residue ID']
merged_df['Residue ID (chain2)'] = chain2_N1_N3_df_reversed['Residue ID']
# Create a new DataFrame with selected columns from merged_df
new_df = merged_df[['atom_i', 'atom_j', 'func', 'r0', 'r1', 'r2', 'k(kJ mol^-1 nm^-2)',";",'Atom ID (chain1)','Atom ID (chain2)','Residue Name (chain1)','Residue Name (chain2)','Residue ID (chain1)','Residue ID (chain2)' ]]

new_df

```

```python
# Open the topology file in append mode
with open('create_topology_file.itp', 'a') as topo_file:
    # Write the [ bonds ] section header
    topo_file.write('\n[ bonds ]\n')
    
    # Iterate through new_df and write each row in the desired format
    for _, row in new_df.iterrows():
        topo_file.write(f"{row['atom_i']:>5} {row['atom_j']:>5} {row['func']:>5} {row['r0']:>5} {row['r1']:>8.3f} {row['r2']:>8.3f} {row['k(kJ mol^-1 nm^-2)']:>8} {row[';']} {row['Residue Name (chain1)']} {row['Residue Name (chain2)']} {row['Residue ID (chain1)']} {row['Residue ID (chain2)']}\n")
```

```python
import MDAnalysis as mda
import pandas as pd
u=mda.Universe("DMPC_DMTAP_single_O1P_O2P_clean_DNA_new1_box_min.gro")
P_atoms = u.select_atoms("resname DA DG DT DC HFA HFG HSA HSG HST HFT HSC HFC and name P")
print(f"{len(P_atoms)}")
# Convert P_atoms to a DataFrame
P_atoms_df = pd.DataFrame({
    'Atom ID': P_atoms.ids,
    'Atom Name': P_atoms.names,
    'Residue Name': P_atoms.resnames,
    'Residue ID': P_atoms.resids})
# print(P_atoms_df.tail(2))


DNA_fix = P_atoms_df[P_atoms_df['Residue ID'].isin([3, 292, 40, 255,71, 224, 113, 182 ])]
print(DNA_fix) 


n = len(DNA_fix)

rest = pd.DataFrame({
    'i': DNA_fix['Atom ID'],
    'func': [2] * n,
    'g': [5] * n,
    'r': [3] * n,
    'f': [1000] * n,
})

# --------------------------------------------------
# 6. Add per-line comments (for humans)
# --------------------------------------------------
rest['comment'] = (
    DNA_fix['Residue Name'] +
    ' resid ' +
    DNA_fix['Residue ID'].astype(str) +
    ' ' +
    DNA_fix['Atom Name']
)

# --------------------------------------------------
# 7. Write GROMACS POSRES file
# --------------------------------------------------
with open("posre_fixDNA.itp", "w") as f:
    f.write("#ifdef POSRESFIXDNA\n")
    f.write("; i  funct  g  r(nm)  k(kJ mol^-1 nm^-2)\n")
    f.write("[ position_restraints ]\n")

    for _, row in rest.iterrows():
        f.write(
            f"{int(row['i']):>6} "
            f"{int(row['func']):>2} "
            f"{int(row['g']):>2} "
            f"{int(row['r']):>3} "
            f"{int(row['f']):>6} "
            f"; {row['comment']}\n"
        )

    f.write("#endif\n")

print("posre_fixDNA.itp written successfully")
```
