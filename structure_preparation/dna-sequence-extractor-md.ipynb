```python
import MDAnalysis as mda

u = mda.Universe("single_O1P_O2P_310K_dmpc411_dmtap75_486_50ns-200ns.part0002.gro")

dna = u.select_atoms("resname DA DT DG DC HFA HFG HFT HFC HSA HSG HST HSC")

modified = {"HFA", "HFG", "HFT", "HFC", "HSA", "HSG", "HST", "HSC"}

sequence = []

for res in dna.residues:
    base = res.resname[-1]   # A/T/G/C

    if res.resname in modified:
        base = "*" + base

    sequence.append(base)

print("Number of residues =", len(sequence))   # should be 294

# split by residue count, not string length
seq1 = sequence[:147]
seq2 = sequence[147:][::-1]

print("seq1 length =", len(seq1))
print("seq2 length =", len(seq2))

print("seq1 =5'", "".join(seq1))
print("seq2 =3'", "".join(seq2))
```
